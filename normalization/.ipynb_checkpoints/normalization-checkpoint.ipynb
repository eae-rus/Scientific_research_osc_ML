```python
import time
from tqdm.auto import tqdm
```

```python
import numpy as np
```

```python
path = '../../datasets/pqd/raw/f_network = 50 and f_rate = 1600/'
```

```python
osc_files = sorted([file for file in os.listdir(path)
                          if "cfg" in file], reverse=False)
```

```python
len(osc_files)
```

```python
#comtrade.load_as_dataframe(path + '0147acceb145f2b6183e3e4c3d5a8a9a.cfg')
```

```python
window_size = 32
step_size = 1
unread = []
for file in tqdm(osc_files):
    try:
        osc_df = comtrade.load_as_dataframe(path + file)
    except:
        unread.append(file)
        continue
    osc_columns = osc_df.columns
    osc_features = []
    to_drop = []
    for column in osc_columns:
        if column in features:
            osc_features.append(column)
        else:
            to_drop.append(column)
    osc_df.drop(columns=to_drop, inplace=True)
    osc_fft = np.fft.fft(osc_df.iloc[:window_size],axis=0).real
    for i in range(window_size, osc_df.shape[0], step_size):
        window_fft = np.fft.fft(osc_df.iloc[i - window_size: i],axis=0).real
        osc_fft = np.maximum(osc_fft, window_fft)
```
