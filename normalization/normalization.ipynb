```python
def analyze_u(file, h1_df, hx_df):
    t1 = 20 * np.sqrt(2)
    t2 = 140 * np.sqrt(2)
    t3 = 560 * np.sqrt(2)
    features = h1_df.columns
    df = pd.DataFrame({'name': [file[:-4]]})
    df['norm'] = 'ДА'
    for r in ru_cols:
        c = list(u_cols[r].intersection(features))
        print(c)
        m1 = h1_df[c].max(axis=1)[0]
        mx = hx_df[c].max(axis=1)[0]
        if np.isnan(m1):
            continue
        df[r + '_h1'] = m1
        df[r + '_hx'] = mx
        if m1 <= t1:
            if m1 <= 1.5 * mx:
                df[r + '_PS'] = 's'
                df[r + '_base'] = 'шум'
            else:
                df[r + '_PS'] = 'p'
                df[r + '_base'] = '?1'#None
        elif t1 < m1 <= t2:
            if m1 <= 1.5 * mx:
                df[r + '_PS'] = 's'
                df[r + '_base'] = '?2'
            else:
                df[r + '_PS'] = 's'
                df[r + '_base'] = 100
        elif t2 < m1 <= t3:
            if m1 <= 1.5 * mx:
                df[r + '_PS'] = 's'
                df[r + '_base'] = '?2'
            else:
                df[r + '_PS'] = 's'
                df[r + '_base'] = 400
        else:
            df[r + '_PS'] = '?3'
            df[r + '_base'] = '?3'
    t1 = 0.03 * np.sqrt(2)
    t2 = 30 * np.sqrt(2)
    for r in ri_cols:
        c = list(i_cols[r].intersection(features))
        m1 = h1_df[c].max(axis=1)[0]
        mx = hx_df[c].max(axis=1)[0]
        print(m1,mx)
        if np.isnan(m1):
            continue
        df[r + '_h1'] = m1
        df[r + '_hx'] = mx
        if m1 <= t1:
            if m1 <= 1.5 * mx:
                df[r + '_PS'] = 's'
                df[r + '_base'] = 'шум'
            else:
                df[r + '_PS'] = 'p'
                df[r + '_base'] = '?1'
        elif t1 < m1 <= t2:
            if m1 <= 1.5 * mx:
                df[r + '_PS'] = 's'
                df[r + '_base'] = '?2'
            else:
                df[r + '_PS'] = 's'
                df[r + '_base'] = 5
        else:
            df[r + '_PS'] = '?3'
            df[r + '_base'] = '?3'
            
    if list(hz_cols.intersection(features)):
        t1 = 0.02 * np.sqrt(2)
        t2 = 5 * np.sqrt(2)
        c = list(hz_cols.intersection(features))
        for ci in c:
            m1 = h1_df[ci][0]
            mx = hx_df[ci][0]
            df[ci[8] + 'Iz_h1'] = m1
            df[ci[8] + 'Iz_hx'] = mx
            if m1 <= t1 or mx <= t1:
                df[ci[8] + 'Iz_PS'] = 's'
                df[ci[8] + 'Iz_base'] = 'шум'
            elif t1 < m1 <= t2 or t1 < mx <= t2:
                df[ci[8] + 'Iz_PS'] = 's'
                df[ci[8] + 'Iz_base'] = '1'
            else:
                df[ci[8] + 'Iz_PS'] = '?3'
                df[ci[8] + 'Iz_base'] = '?3'
                
    dsig = list(d_cols.intersection(features))
    if dsig:
        m1 = h1_df[dsig].max(axis=1)[0]
        df['dId_h1'] = m1
        if m1 < 30:
            df['dId_PS'] = 's'
            df['dId_base'] = 1
        else:
            df['dId_PS'] = '?3'
            df['dId_base'] = '?3'
    if 'шум' in df.values:
        df['norm'] = 'хз'
    if '?1' in df.values or '?2' in df.values or '?3' in df.values:
        df['norm'] = 'НЕТ'
    if list(raw_cols.intersection(features)):
        df['norm'] = 'raw'
    if file[:-4] in raw_files:
        df['norm'] = 'raw'
    return df
```
