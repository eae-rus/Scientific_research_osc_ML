```python
window_size = 32
step_size = 1
unread = []
rec = comtrade.load("../../datasets/pqd/raw/f_network = 50 and f_rate = 1600/f758347852450afce8b2b2662351b0ac.cfg")
osc_df = rec.to_dataframe()
#osc_df = comtrade.load_as_dataframe('../../datasets/pqd/test/913f9526b55d822a6e47650b3afa244a.cfg')
osc_columns = osc_df.columns
osc_features = []
to_drop = []
for column in osc_columns:
    if column in features:
        osc_features.append(column)
    else:
        to_drop.append(column)
osc_df.drop(columns=to_drop, inplace=True)
osc_fft = np.abs(np.fft.fft(osc_df.iloc[:window_size],axis=0))
for i in range(window_size, osc_df.shape[0], step_size):
    window_fft = np.abs(np.fft.fft(osc_df.iloc[i - window_size: i],axis=0))
    osc_fft = np.maximum(osc_fft, window_fft)
h1 = osc_fft[1]
hx = np.max(osc_fft[2:16], axis=0)
h1_df = pd.DataFrame([dict(zip(osc_features, h1))])
hx_df = pd.DataFrame([dict(zip(osc_features, hx))])
```
